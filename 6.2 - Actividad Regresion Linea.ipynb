**Ventas de mi negocio**

Nuestro objetivo será contruir un modelo que prediga las ventas del producto dependiendo de cuánto presupuesto se gaste en cada medio.

1. Carga los datos del archivo AdvmiNegocio.csv

```python
from google.colab import drive
drive.mount('/content/drive/')

# Nos cambiamos a la carpeta donde tengamos el repositorio
%cd 'drive/MyDrive/SemanaTec/Repos/arte-analitica'


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

negocio_df = pd.read_csv('data/AdvmiNegocio.csv', )
```

2. Generar una grafica para visualizar  cómo se relaciona cada columna de presupuesto contra las ventas. De mi negocio. ( Redes,Youtube,TV)


```python
# La primera columna indica el número de renglón. No la vamos a ocupar.
negocio = negocio_df.iloc[:, 1::]
negocio.head()

# Vamos a ver cómo se relaciona cada columna de presupuesto contra las ventas.
# Grafiquemos en una misma figura las tres relaciones.
fig, axes = plt.subplots(1, 3, figsize=(20,6))
fig.suptitle('Presupuesto y Ventas')
# axs son nuestros espacios para graficar. Tenemos 3 espacios.

# Redes
#axs[0] nos dice que vamos a graficar en el primer espacio.
sns.scatterplot(data=negocio, x='Redes',y='sales', ax=axes[0])
axes[0].set_title('Redes')
axes[0].set_ylabel('Ventas ($)')
axes[0].set_xlabel('Cantidad')

# Youtube
#axs[1] nos dice que vamos a graficar en el segundo espacio.
sns.scatterplot(data=negocio, x='Youtube',y='sales', ax=axes[1], color='red')
axes[1].set_title('Youtube')
axes[1].set_ylabel('Ventas ($)')
axes[1].set_xlabel('Cantidad')

# TV
sns.scatterplot(data=negocio, x='TV',y='sales', ax=axes[2], color='green')
axes[2].set_title('TV')
axes[2].set_ylabel('Ventas ($)')
axes[2].set_xlabel('Cantidad')
```

 A partir de los gráficos superiores, se puede constatar que los datos de todos los modelos siguen una tendencia de proporcionalidad positiva lo que quiere decir que conforme aumentan la cantidad de anuncios, aumentarán consecuentemente las ventas. 4. **Regresión lineal**. En su forma más simple, consiste en asumir que una variable $x$ y una variable $y$ presentan una relación lineal de la forma:

$$ y \approx \beta_0 + \beta_1 \cdot x$$

¿Cuál es el valor de beta_0 , beta_1 y el Score  para Redes vs Ventas, TV vs Ventas y YouTube vs Ventas?

```python
# Con esta función, podemos calcular los coeficientes para cada una de las variables

# Televisión
lm1 = LinearRegression()
# La variable X la ponemos con doble corchete!
lm1.fit(negocio[['Redes']], negocio['sales'])
print('Redes')
print('beta_0: ', lm1.intercept_)
print('beta_1: ', lm1.coef_)
print('Score: ', lm1.score(negocio[['Redes']], negocio['sales']), '\n')

# Radio
lm2 = LinearRegression()
# La variable X la ponemos con doble corchete!
lm2.fit(negocio[['Youtube']], negocio['sales'])
print('Youtube')
print('beta_0: ', lm2.intercept_)
print('beta_1: ', lm2.coef_)
print('Score: ', lm2.score(negocio[['Youtube']], negocio['sales']), '\n')


# Periódico
lm3 = LinearRegression()
# La variable X la ponemos con doble corchete!
lm3.fit(negocio[['TV']], negocio['sales'])
print('TV')
print('beta_0: ', lm3.intercept_)
print('beta_1: ', lm3.coef_)
print('Score: ', lm3.score(negocio[['TV']], negocio['sales']))
```

6.  **Regresión lineal múltiple**. Como ahora vamos a incluir más de una variable, el modelo se rescribe a :

$$ \hat{y} = \hat{\beta}_0 + \hat{\beta}_1 \cdot x_1 + \hat{\beta}_2 \cdot x_2 + \hat{\beta}_3\cdot x_3 + \ldots \epsilon $$

Elige  las diferentes combinaciones de variables. Ejemplo: Redes y Youtube vs Ventas 

¿Cuál es el valor de beta_0, betas_1 , beta_2 , el score y el score ajustado?



```python
# Vamos a calcular el modelo con dos variables simultáneas
# Televisión  y radio
lm4 = LinearRegression()
# La variable X la ponemos con doble corchete!
lm4.fit(negocio[['Redes', 'Youtube']], negocio['sales'])
print('Redes y Youtube')
print('beta_0: ', lm4.intercept_)
print('betas: ', lm4.coef_)
print('Score: ', lm4.score(negocio[['Redes', 'Youtube']], negocio['sales']), '\n')

# Tv y periódico
lm5 = LinearRegression()
# La variable X la ponemos con doble corchete!
lm5.fit(negocio[['Redes','TV']], negocio['sales'])
print('Redes y TV')
print('beta_0: ', lm5.intercept_)
print('betas: ', lm5.coef_)
print('Score: ', lm5.score(negocio[['Redes','TV']], negocio['sales']), '\n')


# Periódico y radio
lm6 = LinearRegression()
# La variable X la ponemos con doble corchete!
lm6.fit(negocio[['TV','Youtube']], negocio['sales'])
print('TV y Youtube')
print('beta_0: ', lm6.intercept_)
print('betas: ', lm6.coef_)
print('Score: ', lm6.score(negocio[['TV','Youtube']], negocio['sales']))
```
